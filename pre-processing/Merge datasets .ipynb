```python
import pandas as pd
```

```python
bureau_df = pd.read_csv("bureau.csv")
bureau_df.head()
bureau_df = bureau_df[bureau_df["AMT_ANNUITY"].notna()]
```

```python
bureau_balance_df = pd.read_csv("bureau_balance.csv")
bureau_balance_df.head()
```

```python
bureau_final = bureau_df.merge(bureau_balance_df, on = "SK_ID_BUREAU", how ="left")
```

```python
bureau_final.shape
```

```python
bureau_final.isna().sum()
```

```python
bureau_final= bureau_final[bureau_final["MONTHS_BALANCE"].notna()]
bureau_final= bureau_final[bureau_final["STATUS"].notna()]
bureau_final.isna().sum()
```

```python
bureau_final.drop(['AMT_CREDIT_MAX_OVERDUE'], axis=1, inplace = True)
```

```python
bureau_final.loc[bureau_final['AMT_CREDIT_SUM_LIMIT'].isna(), 'AMT_CREDIT_SUM_LIMIT'] = bureau_final['AMT_CREDIT_SUM_LIMIT'].mean()
```

```python
bureau_final.loc[bureau_final['AMT_CREDIT_SUM_DEBT'].isna(), 'AMT_CREDIT_SUM_DEBT'] = bureau_final['AMT_CREDIT_SUM_DEBT'].mean()
```

```python
bureau_final.loc[bureau_final['DAYS_ENDDATE_FACT'].isna(), 'DAYS_ENDDATE_FACT'] = bureau_final['DAYS_ENDDATE_FACT'].mean()
```

```python
bureau_final.loc[bureau_final['DAYS_CREDIT_ENDDATE'].isna(), 'DAYS_CREDIT_ENDDATE'] = bureau_final['DAYS_CREDIT_ENDDATE'].mean()
```

```python
bureau_final.isna().sum()
```

```python
bureau_final.dtypes
```

```python
bureau_final_ohe = pd.get_dummies(bureau_final, columns = ['CREDIT_ACTIVE', 'CREDIT_CURRENCY', 'CREDIT_TYPE', 'STATUS'])
bureau_final_ohe.head()
```

```python
bureau_final_ohe.to_csv("bureau_final_ohe.csv", index = False)
```

```python
final_df = pd.read_csv("final.csv")
final_df.head()
```

```python
final_df['SK_ID_CURR'].nunique()
```

```python
bureau_final_ohe['SK_ID_CURR'].nunique()
```

```python
from tqdm.auto import tqdm
tqdm.pandas()
def preprocess(x):
    df = pd.merge(final_df, x, on=['SK_ID_CURR'], how='left')
    df.to_csv("final_merged_small_l.csv", mode="a", index=False)

reader = pd.read_csv("bureau_final_ohe.csv", chunksize=10000)

for r in tqdm(reader):
    preprocess(r)
```
